#Astrologer Earning

```python
import re
astro_earning= df[(df["astrologerCallStatus"]=="completed") & (df["callStatus"]=="completed")]
astro_earning['guruName'] = astro_earning['guruName'].str.strip().str.lower()
astro_earning['guruName'] = astro_earning['guruName'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
astro_earning=astro_earning.groupby("guruName")["astrologersEarnings"].apply(np.sum)
astro_earning=astro_earning.sort_values(ascending=False).reset_index()
astro_earning.head(10)
```

These are top astrologer eraning Dr Balakrishna holds 1st postion in earning

#Which astrologer spent more time on call

```python
astro_on_call= df[df["astrologerCallStatus"]=="completed"]
astro_on_call['guruName'] = astro_on_call['guruName'].str.strip().str.lower()
astro_on_call['guruName'] = astro_on_call['guruName'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
astro_on_call=astro_on_call.groupby("guruName")["astrologerOnCallDuration"].apply(np.sum)
astro_on_call=astro_on_call.sort_values(ascending=False).reset_index()
astro_on_call.head(10)
```

Astro Shalini has worked hard and spent more times with customer and also Shalini has more number of failed chat status and incomplete chat status, one possible reason is communication skills and 2nd problem may be her less knwledge on topic

```python
merged_astro= astro_on_call.merge(astro_earning, on="guruName",how="inner")
merged_astro.head(3)
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=merged_astro, x="astrologerOnCallDuration",y="astrologersEarnings")
plt.show()
```

represents that the astrologer who spent more time is not guaranteed that they will earn more
